```python
import pandas as pd
```

```python
pred=pd.read_csv('s3://presidencia-mi-salud/processed/loo_67.csv')
```

```python
pred[['0', '1', '2', '3', '4', '5', '6', '7', '8', 'pred_loo', 'categ_opi']].head()
```

```python
pred['conc']=((pred[['0', '1', '2', '3', '4', '5', '6', '7', '8']]*100)*(pred[['0', '1', '2', '3', '4', '5', '6', '7', '8']]*100)).sum(axis=1)
```

```python
pred=pred.sort_values('conc', ascending=False)
```

```python
pred[['texto', '0', '1', '2', '3', '4', '5', '6', '7', '8', 'pred_loo', 'categ_opi']].tail(20)
```

```python
pred.categ_opi.value_counts()
```

```python
pred['0']=pred['0']/(pred.categ_opi.value_counts()/pred.shape[0])['emergencia']
pred['1']=pred['1']/(pred.categ_opi.value_counts()/pred.shape[0])['informacion']
pred['2']=pred['2']/(pred.categ_opi.value_counts()/pred.shape[0])['nacimiento']
pred['3']=pred['3']/(pred.categ_opi.value_counts()/pred.shape[0])['otra']
pred['4']=pred['4']/(pred.categ_opi.value_counts()/pred.shape[0])['otra_queja']
pred['5']=pred['5']/(pred.categ_opi.value_counts()/pred.shape[0])['pregunta']
pred['6']=pred['6']/(pred.categ_opi.value_counts()/pred.shape[0])['pregunta_busca trabajo']
pred['7']=pred['7']/(pred.categ_opi.value_counts()/pred.shape[0])['pregunta_medica']
pred['8']=pred['8']/(pred.categ_opi.value_counts()/pred.shape[0])['respuesta']
```

```python
pred[['0', '1', '2', '3', '4', '5', '6', '7', '8', 'pred_loo', 'categ_opi', 'texto']].sort_values('8',ascending=False)
```

```python
pred['pred_ponde']=pred[['0', '1', '2', '3', '4', '5', '6', '7', '8']].idxmax(axis=1)
```

```python
label_map={'0':'emergencia' ,
          '1': 'informacion',
          '2': 'nacimiento',
          '3': 'otra',
          '4': 'otra_queja',
          '5': 'pregunta',
          '6': 'pregunta_busca trabajo',
          '7': 'pregunta_medica',
          '8': 'respuesta'}

pred.pred_ponde=pred.pred_ponde.map(label_map)
```

```python
pd.crosstab(pred['pred_ponde'], pred['pred_loo'])
```

```python
sum(pred['pred_loo']==pred['categ_opi'])/pred.shape[0]
```

```python
sum(pred['pred_ponde']==pred['categ_opi'])/pred.shape[0]
```

```python
from sklearn.metrics import classification_report
```

```python
pred['pred_ponde_mod']=pred['pred_ponde']

pred['pred_ponde_mod'].loc[pred['pred_ponde_mod']=='otra_queja']='otra'
pred['pred_ponde_mod'].loc[pred['pred_ponde_mod']=='pregunta_busca trabajo']='pregunta'
pred['pred_ponde_mod'].loc[pred['pred_ponde_mod']=='pregunta_medica']='pregunta'


pred['pred_mod']=pred['pred_loo']


pred['pred_mod'].loc[pred['pred_mod']=='otra_queja']='otra'
pred['pred_mod'].loc[pred['pred_mod']=='pregunta_busca trabajo']='pregunta'
pred['pred_mod'].loc[pred['pred_mod']=='pregunta_medica']='pregunta'

pred['categ_opi_mod']=pred['categ_opi']


pred['categ_opi_mod'].loc[pred['categ_opi_mod']=='otra_queja']='otra'
pred['categ_opi_mod'].loc[pred['categ_opi_mod']=='pregunta_busca trabajo']='pregunta'
pred['categ_opi_mod'].loc[pred['categ_opi_mod']=='pregunta_medica']='pregunta'
```

```python
print(recall_score(y_pred=pred['pred_mod'],
                      y_true= pred['categ_opi_mod'], average='macro' ))


print(recall_score(y_pred= pred['pred_ponde_mod'],y_true= pred['categ_opi_mod'], average='macro' ))

```

```python
pred['conc_bins']=pd.qcut(pred.conc, 100,
        retbins=False, precision=0, duplicates='raise')

pred.conc_bins=pred.conc_bins.factorize()[0]
```

```python
for i in range(1, 100):
    print(classification_report(y_pred= pred['pred_ponde_mod'].loc[pred.conc_bins.isin(range(0, i))],
                                y_true= pred['categ_opi_mod'].loc[pred.conc_bins.isin(range(0, i))] ))
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
recall_score(average='macro', y_pred= pred['pred_ponde_mod'].loc[pred.conc_bins.isin(range(0, i))],
                                y_true= pred['categ_opi_mod'].loc[pred.conc_bins.isin(range(0, i))])
```

```python
precision_score(average='macro', y_pred= pred['pred_ponde_mod'].loc[pred.conc_bins.isin(range(0, i))],
                                y_true= pred['categ_opi_mod'].loc[pred.conc_bins.isin(range(0, i))])
```

```python
rango=pd.DataFrame()
for i in range(1, 100):
    dec=i
    prec=precision_score(average='macro', y_pred= pred['pred_ponde_mod'].loc[pred.conc_bins.isin(range(0, i))],
                                y_true= pred['categ_opi_mod'].loc[pred.conc_bins.isin(range(0, i))])

    rec=recall_score(average='macro', y_pred= pred['pred_ponde_mod'].loc[pred.conc_bins.isin(range(0, i))],
                                y_true= pred['categ_opi_mod'].loc[pred.conc_bins.isin(range(0, i))])
    
    n_obs=pred.loc[pred.conc_bins.isin(range(0, i))].shape[0]
    
    rango_i=pd.DataFrame({'prec':prec,
                          'rec':rec,
                          'dec':dec,
                          'n_obs':n_obs}, index=[i])
    
    rango=pd.concat([rango, rango_i])
```
