```python
%%timeit
data_fin=pd.DataFrame()
for i in range(0, 1): 
    contact_uuid = data.loc[i, 'contact_uuid']
#    print('haciendo peticion...')
    response = requests.get("http://10.20.55.67/api/v2/runs.json",
                            headers={"Authorization": "Token 0032fe79dbddceae3f4a86e86a726e16b88ec88e"},
                            params={"contact": contact_uuid})
#    print('formateando dataframe')
    response=json.loads(response.text)
    ultimo_tmstmp=response['results'][0]['created_on']
    penultimo_tmstmp=response['results'][1]['created_on']
    flujo_penultimo=response['results'][1]['flow']['name']
    flujo_ultimo=response['results'][1]['flow']['name']
    

    out=pd.DataFrame({'contact_uuid':data.loc[i, 'contact_uuid'],
                  'nombre':data.loc[i, 'nombre'],
                  'texto':data.loc[i, 'texto'],
                  'ultimo_tmstmp':ultimo_tmstmp,
                  'penultimo_tmstmp':penultimo_tmstmp,
                  'flujo_penultimo':flujo_penultimo,
                  'flujo_ultimo':flujo_ultimo}, index=[i]
    )
#    print('OK! '+ str(i))
    data_fin=pd.concat([data_fin, out], ignore_index=True)

```
